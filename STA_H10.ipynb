Q1.1
Simple linear regression analyzes the relationship between a single independent variable, X, and a dependent variable, Y, with a line of best-fit.
Multiple linear regression accounts for many independent variables and one dependent variable, and finds constants such that Y= b0 +X1b1 + ... +Xnbn + e, where e is a (hopefully) small error term.

Multiple linear regression is "better" because it can handle confounding variables, and can describe how Y changes when one or more of the independent variables are fixed. This means it can more accurately describe real-world relationships, which often have confounding variables.

Q1.2 A continuous variable X in SLR can take on a range of values, so the regression Y = b + b1X represents a line.
An indicator variable is binary, every data point is either in X or not, so Y = b + b1X means that if X=0, the average value of Y is b, and if X=1, the average value of Y is b+b1. So, b1 is how much the data in X differs from the data not in X, on average.

Q1.3 When we introduce an indicator variable alongside a continuous one, the MLR equation Y = b0 + b1X + b2D, for continuous X and binary D, captures two different regressions, depending on if D=0 or D=1.
If D=0, the intercept is b0, and if D=1, the intercept is b0+b2. So b2 is the amount our regression shifts by, depending on if D is 1 or 0.

Q1.4 When we allow for an interaction between our continuous and indicator variables, our linear form reflects this by allowing for a change in the slope of our regression, depending on if we are in the baseline group or not.
Our linear form would look like Y = b0 + b1X + b2D + b3DX. Thus, if D=1 the intercept is adjusted by b2, and the slope is adjusted by b3. This helps model real-world problems since grouping the data often has an effect on the correlation (slope).

Q1.5 If we perform an MLR using binary indicators based on one non-binary categorical variable, our linear form will look like Y=b0+b1X1+...+bnXn, where each of the Xi are binary variables, representing whether a data point satisfies a given category, where the categorical variable has n+1 categories. Each of the b1,...,bn are adjustments from the average of the baseline (b0).

Q2.1 The outcome would be revenue. The inputs are amount spent on TV ads, and amount spent on online ads. You could potentially be more specific, but these should be good enough. Without interaction, we could model the outcome as such: R = b0+b1X+b2Y, where X and Y are TV and online, respectively.
With interactions, we lose the linearity of our model, getting R=b0+b1X+b2Y+b3XY, since the coefficients for X and Y may depend on the other variable.

Q2.2 If the advertising spending variables were instead binary, we would have R=b0+b1X+b2Y, and every possible combination of 1's and 0's is accounted for.

In this session, we explored multiple aspects of linear regression models, particularly focusing on **Multiple Linear Regression (MLR)** and evaluating model performance. Below is a breakdown of the key topics covered.

### **1. Multiple Linear Regression (MLR) vs. Simple Linear Regression (SLR)**

- **SLR** involves modeling a dependent variable based on a single predictor variable, while **MLR** expands on this by using multiple predictor variables.
- **MLR** allows for the inclusion of **interaction terms** (e.g., the combined effect of two variables), **categorical variables** (via encoding), and **continuous variables**, enabling more complex and accurate models.

---

### **3. Evaluation: In-sample and Out-of-sample Performance**

#### **In-Sample Performance (R²)**

- In-sample \( R^2 \) indicates how well the model fits the **training data**. A higher value suggests that the model explains most of the variability in the training data. However, this can be misleading if the model overfits (i.e., learns noise in the training data rather than true patterns).

#### **Out-of-Sample Performance (R²)**

- Out-of-sample \( R^2 \) evaluates the model's performance on **new data**, reflecting its ability to **generalize** to unseen samples.
- A discrepancy where in-sample \( R^2 \) is much higher than out-of-sample \( R^2 \) can indicate overfitting—where the model is too complex and tailored to the training data, reducing its ability to predict unseen data accurately.

---

### **5. Model Interpretation: Multicollinearity**

Throughout the development of these models, we observed that adding numerous interaction terms and including both continuous and categorical variables can lead to **multicollinearity**. This occurs when predictor variables are highly correlated with each other, which can distort the coefficients and affect the model’s ability to generalize to new data.

- For example, the addition of many highly correlated variables can inflate the **condition number** (as observed in the models), which is a sign of multicollinearity.
- **Centering and scaling** help alleviate some of the problems associated with multicollinearity by making the model more stable and improving convergence.
