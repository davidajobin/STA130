### Summary of Key Concepts:

1. **Quantitative vs. Qualitative Data**:
   - **Quantitative Data**: Numeric data that can be measured (e.g., height, weight). It can be continuous or discrete.
   - **Qualitative Data**: Categorical data that describes characteristics or qualities (e.g., colors, names).

2. **Python Libraries**:
   - **Plotly**: Used for creating interactive plots such as bar graphs and histograms.
   - **NumPy**: Contains the `randn()` function, which generates samples from a standard normal distribution.

3. **`randn()` Function**:
   - Generates random samples from a normal distribution with a mean of 0 and a standard deviation of 1.

4. **Histograms in Plotly**:
   - The `nbins` parameter controls the maximum number of bins, but the exact number displayed can be less, depending on data distribution.

5. **Box Plots vs. Bar Plots**:
   - **Box Plot**: Shows data distribution, including median, quartiles, and outliers. Useful for visualizing data spread.
   - **Bar Plot**: Compares quantities across categories, representing data as bars. Does not show data distribution.

6. **Outliers in Box Plots**:
   - Defined as points that fall outside the whiskers, calculated as \( Q1 - 1.5 \times IQR \) (lower bound) and \( Q3 + 1.5 \times IQR \) (upper bound).

7. **Extracting Columns from DataFrames**:
   - **Dot Notation** (`df.column`): Concise but limited to valid Python identifiers.
   - **Bracket Notation** (`df["column"]`): More flexible, works with any column name, including those with spaces or special characters.

8. **Histogram's `color` Variable**:
   - Differentiates subsets of data by coloring bars based on categories (for categorical variables) or gradients (for continuous variables).

9. **Box Plot's `points` Variable**:
   - Controls how data points are displayed:
     - `"outliers"`: Shows only outliers.
     - `"all"`: Displays all data points.
     - `False`: Hides all points.
     - `"suspectedoutliers"`: Shows only suspected outliers.
